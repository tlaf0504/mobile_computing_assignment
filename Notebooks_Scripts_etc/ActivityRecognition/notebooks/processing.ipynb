```python
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
%matplotlib

import utils
import pandas as pd
import seaborn as sns
```

```python
data_directory = os.path.join("..", "data")
```

```python
activities = [
     {"name": "Triceps",
      "dir": "triceps",
      "label" : "triceps"},
     {"name": "Russian Twist",
      "dir": "russian_twist",
      "label" : "russian_twist"},
     {"name": "Curls",
      "dir": "curls",
      "label" : "curls"},
     {"name": "Crunches",
      "dir": "crunches",
      "label" : "crunches"}   
]
```

```python
labeled_data = []
```

```python
def process_activity_data(data):
    m = np.mean(data[:,1:], 0)
    signals_without_mean = data[:,1:] - m
    #fig = utils.plot_sensor_data(data, "tmp", ylabels=["x", "y", "z"])
    #_ = utils.plot_sensor_data(np.column_stack((data[:,0], signals_without_mean)), title="tmp", ylabels=["x", "y", "z"], fig=fig)
    energies = np.var(signals_without_mean, axis=0)
    return energies
```

```python
for activity in activities:
    activity_directory = activity["dir"]
    label = activity["label"]
    
    activity_path = os.path.join(data_directory, activity_directory, "clipped")
    gyro_data_files = glob.glob(os.path.join(activity_path, "*_gyro_sensor_data_*.npy"))
    accel_data_files = glob.glob(os.path.join(activity_path, "*_accel_sensor_data_*.npy"))
    
    gyro_energies = []
    accel_energies = []
    
    for file in gyro_data_files:
        gyro_data = utils.load_and_resample_data(file, src_type="npy")
        gyro_energies.append(process_activity_data(gyro_data))
        
    for file in accel_data_files:
        accel_data = utils.load_and_resample_data(file, src_type="npy")
        accel_energies.append(process_activity_data(accel_data))
        
    for k in range(len(gyro_energies)):
        labeled_data.append([gyro_energies[k][0], gyro_energies[k][1], gyro_energies[k][2], accel_energies[k][0], accel_energies[k][1], accel_energies[k][2], label])
```

```python
data = pd.DataFrame(data = labeled_data,
                    columns=['gyro_x', 'gyro_y', 'gyro_z', 'accel_x', 'accel_y', 'accel_z', 'activity'])
data
```

```python
sns.set_theme()
sns.set_palette(sns.color_palette())
sns.pairplot(data, hue="activity")
```
